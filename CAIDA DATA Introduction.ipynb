# 理解网络流量数据

网络流量数据研究的方向有：单条流数据被研究、聚合在同一条链路上的流数据的研究、网络流量性能指标数据的研究

网络流量数据研究的主要目的： 研究典型的正常流量数据的特征和非典型的流量特征。正常流量数据的研究，目的是理解网络应用流量如何影响流量工程、网络结构、以及协议、路由器、防火墙以及其他网络组件的设计。

    最后，尽管第2节的其余部分重点讨论了如何使用网络数据来直接解决当今Internet面临的问题，但还有一个重要的间接影响:数据可以用来设计、填充和验证网络仿真和建模。由于仿真和建模创建了一个独立的、可延展性的互联网子集版本，因此它在研究重点研究问题时具有不可思议的价值。然而，如果研究人员希望他们的结论能反映出这些约束条件，他们就必须理解模型与现实世界之间的关系。网络仿真的验证、校核和认证领域是[25]的一个重要领域，但在实践中应用[37]是一个挑战。

### 数据分类

#### 地址空间分配数据集：

包含具有一些特征的Internet地址(例如，响应不同代码的地址，或者看起来是动态的地址，等等)。这个数据集中的IP地址通常不是匿名的，因为它们是通过测量流量来确定的，而不是实际的发送方和接收方通信，因此与特定的个体无关。这些数据可以用来更好地了解Internet拓扑结构和地址使用情况。

一些子数据集

- Internet地址统计调查：在大约两个月的时间里，对每个分配的IP地址进行一次ping，并报告结果。
- Internet地址调查：在分配的Internet地址空间中选择/24前缀的1-2%，并在这些块中多次ping每个地址(大约每11分钟一次)，持续两周。
- IP Address HItlists: 一个hitlist为每个/24 IPv4地址块提供了一个代表，该地址块被判断最有可能回复ping.目标是提供响应性强、完整且稳定的代表:即响应ping和traceroutes的地址(概率很高)，覆盖每一个已分配的IPv4 /24前缀，并且不会随时间变化太多。
- IP Address History:IP地址历史记录包含所有曾经响应过16次IPv4地址普查的IPv4地址，以及每个地址对未来ping的响应可能性的得分。地址历史记录对于为拓扑和可靠性研究定制IP地址的选择非常有用.


#### 其他Internet拓扑数据

- Geolocation data： IP地理位置数据集提供指定IP地址的推断地理位置(纬度/经度)，在某些情况下，用于进行推断的源数据。地理位置数据可以用来了解互联网流量从何而来，以及如何受到当地政策的影响。
- AS-to-Organization maps ：将ASes(从BGP路由)与组织(公司)联系起来 
- Internet Topology data: Internet拓扑数据是由一个试图映射Internet的程序创建的。该程序能够确定哪些路由器能够与其他路由器通信。Internet拓扑数据仅显示Internet核心内部和外部企业边界的路由器连接;它不包含任何可识别信息或企业内部拓扑信息。该数据集可用于蠕虫暴发建模与仿真、蠕虫控制与对策、DDoS攻击僵尸分布、漏洞评估、互联网拓扑与地址分布演化纵向研究、互联网拓扑与地址映射推断等。

#### 流量相关数据

基于IP报头的流量数据:这些数据集由流量数据报头组成，包含匿名源和目标IP地址等信息，以及其他IP和传输(例如TCP、UDP、ICMP、SCTP)报头字段。不包含任何数据包内容。根据具体数据集的不同，这类数据可用于描述典型的Internet流量，或描述流量异常(如DDoS攻击、端口扫描或蠕虫暴发)。

- **general anonymized packet headers**： 来自区域网络接入链路的包头的收集。
- **anonymized attack traffic (packet headers)** ：来自区域网络接入链路的包头的收集（包含已知的ｄｏｓ攻击流量）。
- **artificial attacks over real background traffic**: 在真实的网络背景流量上产生了一系列不同强度的人工攻击数据。

基于flow的流量数据:网络流量可以表示为两个端点之间的流量。此数据集包含流量信息，其中包括各种属性，如源和目标IP地址、源和目标端口、协议类型以及包和字节计数。这些数据可以是由一系列不同的收集工具(如NetFlow、IPFIX和argus或变体)生成的不同格式。这些文件中的IP地址在每个数据集或每个时间间隔的基础上匿名。这些数据集对于诸如网络经济学和会计、网络规划、分析、安全性、拒绝服务攻击、网络监视以及流量可视化等研究都是有用的。

- generalized flows：以Argus格式提供一些流数据。
- long-flows：提供的网络流量数据来源于区域网络访问链路上的报头，其格式旨在支持对短流量和长流量的检查。

#### DNS数据

1. **Public DNS data**: 由来自公共资源(如来自公共DNS服务器)的域名系统数据组成。它与用户无关，也没有隐私限制。
2. **Anonymized DNS data**: 由域名系统数据组成，这些数据不包含个体的识别信息，或者是因为它是匿名的，或者是因为它被聚合到个体查询被隐藏的级别，或者仅仅包含来自测试程序的实验数据(不是个体)。
3. **Limited DNS data**: 该数据集由域名系统数据组成，这些数据不能直接识别个人，但不能与其他数据源相结合。


#### Service Enumeration Data

- **Anycast enumeration datasets**: Active probing information to DNS anycast services such as root DNS. Typically probes are made from many vantage points with the goal to enumerate all anycast nodes in the service. Anycast enumeration datasets are useful to understand the operational status and geographic reach of anycast services and nodes. For detailed information about the dataset, please refer to dataset description page.

- **Google front-ends enumeration and mapping**: Active DNS queries with EDNS-client-subnet allow enumeration of Google front-ends IP addresses. With all the front-ends IP addresses, we use new technique to geolocate the front-ends and clustering them into serving sites. For detailed information about the dataset, please refer to dataset description page.

- **Reverse DNS data**: We collect and provide a crawl of the IPv4 Reverse DNS domain names.

- **Other DNS data**: we have data related to DNS backscatter.

#### Internet Outage Data
Datasets in this category record information about Internet outages–address blocks that become unreachable. Typically outages are inferred from active probing. It may include /24 block-level outages over time, or lists of inferred outages that affect larger parts of the Internet. Outage data can be useful to understand Internet reliability.

For detail of the dataset, please refer to the description page and outages formats.

#### Other types of data
other paper-specific datasets: We have several other datasets specific to papers we have published, including p2p traffic detection, TCP SYNs, etc.

M-Lab旨在通过向消费者提供有关互联网性能的有用信息来推进互联网研究。通过提供免费、开放的互联网测量数据，研究人员、监管机构、倡导团体和普通公众可以更好地了解互联网是如何为他们服务的，以及如何为未来维护和改进它。提供一个开放的、可验证的全球网络性能测量平台，托管全球最大的开放互联网性能数据集，创建可视化和工具，帮助人们理解互联网性能。

### 典型的案例

Neubot 测试数据集：

三种测试类型：

- speedtest：通过发起http请求，测量round-trip time，download speed ，upload speed
- bittorent test：通过发起Bit Torrent 协议请求，测量round-trip time，download speed ，upload speed
- raw test：进行一个原始的持续10S的TCP下载，测试下载速度，并且收集关于TCP 发送端的统计指标。

speedtest收集到的数据记录实例：

- uuid：随机序列号用于标识客户端设备，对时间序列分析有用。
- timestamp：测试执行时候的时间戳，自从1970年1月1号以来的流逝时间。
- real_address：Neubot的ip地址，在服务端看到的ip地址。
- remote_address: 服务端的ip地址。
- connect_time：通过测量连接系统调用完成所需的时间(以秒为单位)来估计RTT。
- download_speed：下载速度（接收到的字节数除以下载时间）Bps。
- upload_speed：上传速度（发送的字节数除以上传时间）Bps。
- latency：通过测量在发送一个小请求和收到一个小的回复的时间的平均值。
- internal_address: 在局域网内部的客服端的ip地址。


数据集分析的维度：


- time：客户机在连接到Internet时定期运行测试。需要考虑测量参数随时间的变化，以便更好地理解连接的情况，并在相同条件下立即将其与其他客户机联系起来。
- location：关于客户的近似物理位置的信息将在理解Internet访问参数中检测到的任何异常值是否是由于位置(例如，在某个区域供应的资源稀缺)或其他(例如，提供者施加的限制)方面发挥作用。这可以通过分析在相同条件下其他客户机的参数行为来检测。
- network metrics：需要分析网络指标本身的值，因为主动测量的信息非常丰富，但是它们可能受到其他基于网络的应用程序的并发使用的影响。
- connectivity type：无线技术在互联网接入方面的广泛使用需要根据连接类型的不同考虑不同的参数，以避免将有线连接的数据与无线连接的数据混合，因为提供这种连接的约束和业务模型有很大的不同。

target：  validate their analytical and simulation models with data from real-world Internet paths。
