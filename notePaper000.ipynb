# 仿真大会总结



## 1. 智能制造的几点认识

1. 老概念新技术的结合：智能技术（新一代信息技术）在制造全生命周期的应用中所涉及的理论、方法、技术和应用。智能制造是制造技术、自动化技术、系统工程与人工智能等学科的相互渗透、互相交织而形成的一门综合技术。

2. 智能制造的发展需要提高与普及想结合，即新型人工智能技术在制造范围内的真正落地。

3. 实现智能制造的几个关注点：


- 新视野下的需求分析，企业的三条线（知识产权-创新管理，价值链-业务管理，资产链-运营管理）中都需要智能技术

![智能制造](./imageUsage/智能制造.png)


- 关注智能产品/智能装备的研发-- 大力发展嵌入式智能系统的应用

- 智能工厂的规划（效益驱动，总体规划，分布实施，终点突破原则）

- 智能制造自动化生产线，自动化技术与制造工艺、工艺技术的巧妙结合

- 制造执行系统与智能排产，智能优化技术在复杂排产问题上的应用

4. 智能制造带来深远的影响：对物理世界的智能感知和认知将广泛应用，深刻改变人类生活方式，社会分工面临巨大的变化，社会治理将从定性到定量有质的提高。


## 2. 高性能计算走向何方

1. 高性能计算机的发展驱动力

- 提升运算速度是计算机发展永恒的主题

评价高性能计算机的主要指标：

峰值运算速度（理论速度）， 持续运算速度， 绿色指标（性能功耗比，单位功耗下的运算能力）

- 科学工程应用是高性能计算机的关键和首要驱动力

科学工程应用需要能量计算，广泛用于天体物理、地球科学、环境科学、航天航空、生命科学、能源科学、材料科学、国防情报等许多极具有挑战性的难题。

- 数据科学、人工智能等新兴应用成为高性能计算机的新引擎

数据科学挑战：4V（volume，variety，velocity，value）特征；规模数据量大；价值密度低；类型多种多样，关联错综复杂。
人工智能挑战：计算需求增加，适应性高，大规模推广，计算特殊，专用体系结构

## 3. 相似计算：提高计算机系统的智商

1. 计算机系统发展的永恒目标：

运行更高效，使用更自然，处理更聪明。


2. 人类思维和机器思维的对应关系：

![人类思维0](./imageUsage/人类思维.png)


人类思维的基本表现形式：判断；概念；推理

**计算机系统的智商：** 计算机硬件和系统软件所具有的智力，是计算机系统设计、研发者赋予计算机系统的，使用者可以测试和直接利用的类人类思维能力。

计算机系统智商发展的几个阶段：

**一级智商：** 0-1字符比较与相同判断，是计算机裸机的类人类思维能力

**二级智商：** __演绎__， 确定性程序是一个封闭系统，对具有数据运行的结果是确定的，其正确性是由程序本身的正确性所保证的，因而具有演绎特征。演绎是装配系统软件确定性程序的计算机系统的类人类思维能力。如：操作系统，编译系统，数据库系统，图形系统

新的发展环境下计算机系统智商的表现：在计算机二级智能的支撑下编制的各种人工智能程序，如概念形成（概念聚类，包括统计与符号），机器推理，专家系统，各种机器学习（归纳，类比，深度学习），神经网络（统计归纳），具有新的类人类思维能力。


3. 人工智能算法下计算机系统表现出来智力：聚类算法与概念形成、数据挖掘、基于案例的学习、神经网络、深度学习

人工智能算法的基本模型：

知识库＜数据、领域知识、常识、元知识＞ +　启发式搜索

![人工智能算法的基本模型](./imageUsage/基本模型.png)

**启发式搜索：** 利用与待解问题的有关信息（启发信息），对搜索路径的走向给予一定的约束或选择的搜索方法。

**常识：** 一般的、普通的知识。一般人应当具有的知识。

**元知识：** 是领域知识和常识层级之上的知识，是控制和操作知识的知识。

4. 相似计算单元的提出

人工智能系统的类人类思维能力是：<font color='red' size=5 face="黑体">归纳、类比</font>

通过对归纳/类比过程的分析，提高计算机系统智商的基本操作包括：**<font color='blue' size=5>相似定义、逼近（相似性比较）、假设生成、模糊/概率计算</font>**

更为通用可以提高计算机系统思维能力的芯片：**相似计算单元 <font color='green' size=4>SPU</font>**

**相似计算的特点：近似、模糊/统计，结果的或然性**

三类计算单元的对比：

- ** U： 系统软件 --->类人类思维：判断、演绎**
- ** U： 并行加速**
- ** U： 知识软件 --->类人类思维：归纳、类比**

SPU ：跨越计算机逻辑思维和非逻辑思维的鸿沟
